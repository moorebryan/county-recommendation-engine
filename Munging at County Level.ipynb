```python
# Function to scrape non-zip files from website and store on local machine
def download_nonZip_file(file_url,file_name):
    file_path = os.path.join(raw_data_folder,file_name)
    # Check to see if file already in the folder
    if os.path.exists(file_path):
        # If it is in the folder delete it before downloading a fresh copy
        os.remove(file_path)
    # Now download a fresh copy from the site
    wget.download(file_url, os.path.join(raw_data_folder,file_name))
```

```python
# Get correct column titles for this dataset
rowLabel=7
wb = open_workbook(os.path.join(raw_data_folder,mean_income_etc_name))
count=0
for sheet in wb.sheets():
    if count==0:
        listofLabels = sheet.row_values(rowLabel)
    count+=1
mean_income_etc_df = pd.read_excel(os.path.join(raw_data_folder,
                                              mean_income_etc_name),encoding = "ISO-8859-1",names=listofLabels)

# Due to extra information at top of this file drop first few rows
mean_income_etc_df.drop(mean_income_etc_df.head(8).index, inplace=True)

# Rename column FIPStxt to FIPS
mean_income_etc_df.rename(index=str, columns={"FIPStxt": "FIPS"}, inplace=True)

# Now look at output
mean_income_etc_df.head()
```

```python
# Add column for FIPS to home_num_of_sales_df
# First we change column name in home_num_of_sales_df to match that in crosswalk_count_df
# We thus change RegionID to CountyRegionID_Zillow, as this is what we will merge on
home_num_of_sales_df.rename(index=str, columns={"RegionID": "CountyRegionID_Zillow"}, inplace=True)

# Now merge the FIPS column from crosswalk_count_df into home_num_of_sales_df by matching CountyRegionID_Zillow
home_num_of_sales_df = pd.merge(home_num_of_sales_df,
                 crosswalk_count_df[['CountyRegionID_Zillow', 'FIPS']],
                 on='CountyRegionID_Zillow')

home_num_of_sales_df.describe()
```

```python
# Add column for FIPS to crime_rate_county_data
# To get this we combine the FIPS for the state and the individual FIPS for the County

crime_rate_county_data_df['FIPS'] = crime_rate_county_data_df.FIPS_ST*1000 + crime_rate_county_data_df.FIPS_CTY

crime_rate_county_data_df.head()
```

```python
# Add column for FIPS to home_median_listing_price_df
# To get this we combine the FIPS for the state and the individual FIPS for the County

home_median_listing_price_df['FIPS'] = home_median_listing_price_df.StateCodeFIPS*1000 + home_median_listing_price_df.MunicipalCodeFIPS

home_median_listing_price_df.head()
```

```python
# Add column for FIPS to home_median_listing_price_per_sqr_ft_df
# To get this we combine the FIPS for the state and the individual FIPS for the County

home_median_listing_price_per_sqr_ft_df['FIPS'] = home_median_listing_price_per_sqr_ft_df.StateCodeFIPS*1000 + home_median_listing_price_per_sqr_ft_df.MunicipalCodeFIPS


home_median_listing_price_per_sqr_ft_df.head()
```

```python
# This will hold our data
county_attr_df = pd.DataFrame()
```
